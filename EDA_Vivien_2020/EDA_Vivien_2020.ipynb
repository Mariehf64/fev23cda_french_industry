```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
dtypes_dic = pd.read_csv('variables_2020_dtypes.csv')
dtypes_dic = dict(zip(dtypes_dic['Variables'], dtypes_dic['Datatypes']))
df = pd.read_csv('../all_content_2020.csv', dtype=dtypes_dic)
df.head()
```

# LISTE DES VARIABLES #

```python
variables = pd.DataFrame({'Variables': df.columns.to_list(), '% de NA': np.round(df.isna().sum() / len(df) * 100,2), 'Datatypes': df.dtypes.values}).reset_index(drop=True)
```

```python
variables.Datatypes = variables.Datatypes.apply(lambda x: 'string' if x == 'object' else x)
```

```python
variables.loc[variables['Variables'] == 'Code_postal', ['Datatypes']] = 'string'
```

```python
variables[['Variables', 'Datatypes']].to_csv('variables_2020_dtypes.csv')
```
