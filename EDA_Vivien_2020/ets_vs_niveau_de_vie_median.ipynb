```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
#chargement du dataset 2020 global
df = pd.read_csv('../all_content_2020_v4.csv', dtype={'CODGEO': 'string'})

#Selection du niveau de vie median et CODGEO pour la jointure
target = pd.DataFrame({'CODGEO': df.CODGEO, 
                    'nv_vie_median': df.MED20})

#Chargement du dataset des établissements
ets = pd.read_csv('variables_ets.csv')
```

```python
len(ets)
```

```python
#jointure ets et target
ets = ets.merge(target, how='left', on='CODGEO')
```

```python
# La colonne du niveau de vie ne contient pas que des floats. Quels sont les termes non numériques que l'on retrouve?
non_float = []
for i in ets.nv_vie_median:
    try:
        float(i)
    except:
        non_float.append(i)
        
print(np.unique(non_float))
```

```python
ets.nv_vie_median.value_counts()
```

```python
ets.nv_vie_median = ets.nv_vie_median.replace(['nd', 's'], [np.nan, np.nan])
```

```python
ets.nv_vie_median = ets.nv_vie_median.astype(float)
```

```python
ets = ets.dropna(subset=['nv_vie_median'], axis=0)
```
