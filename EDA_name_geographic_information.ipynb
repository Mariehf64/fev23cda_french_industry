```python
# importation des librairies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**ANALYSE DE NaN**

aucun nan idenfié via info()



```python
rap_na=pd.DataFrame()
rap_na['Nbre_Na']=df.isna().sum()
rap_na['% NAN']=((df.isna().sum()/df.shape[0])*100).round(2)

display(rap_na)
```

```python
# nombre d'enregistrement un NaN
print("nombre d'enregistrement contenant au moins un NaN : ",df.isna().any(axis=1).sum())

# enregistrement contenant au moins un NaN
display(df[df.isna().any(axis=1)==1])
```

**CHANGEMENT DE DTYPE**

ci-dessous les variables qui devraient être catégorielles :

*   code_région
*   numéro_circonscription
*   code_insee

Ci-dessous la variable qui devrait être continue :

*   longitude

```python
# changement dtype en str 

change_dtype_str={'code_région':'str',
              'numéro_circonscription':'str',
              'code_insee':'str',
              }
df=df.astype(change_dtype_str)

df.info()
```

```python
# changement dtype de la variable 'longitude' en float

# identifier le problème de conversion de la variable 'longitude'
liste=[]

for i in df.longitude :
  try :
    pd.to_numeric(i)

  except ValueError :
    liste.append(i)  
print(liste)  # les entrées problèmatiques
print(len(liste)) # nombre d'erreurs
```

```python
# remplacer "-" par Na
print("nombre de na avant transformation : ",df.longitude.isna().sum())  # nombre de na avant transformation
print("nombre d'enregistrement concernés par le changement : ",df.longitude[df.longitude=='-'].count()) # nombre d'enregistrement concernés par le changement
df.longitude=df.longitude.replace(to_replace='-',value=np.nan) # changement 
print("nombre de na après transformation :",df.longitude.isna().sum()) # nombre de na après transformation
```

```python
 # remplacer les ',' par '.'
df_test=pd.DataFrame()
df_test['longitude']=df.longitude[df.longitude.str.contains(',')==1] 
df_test.longitude=df.longitude.str.replace(',','.')
df_test
```
